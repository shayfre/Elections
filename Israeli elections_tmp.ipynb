```python
excel_file = "22kneset_ballot_boxes.xlsx"
df = pd.read_excel(excel_file)
df.rename(columns={'סמל ועדה':'committee symbol','שם ישוב':'settlement name','סמל ישוב':'settlement symbol','בזב':'suffrages','מצביעים':'voters','פסולים':'disqualified','כשרים':'qualified'}, inplace=True)

df['vote_percentage'] = df['voters'] / df['suffrages'] * 100

df.head(5)
```

```python
print(df.axes[1])
start_idx = df.columns.get_loc('qualified') + 1
end_idx = df.columns.get_loc('vote_percentage')
parties_columns = df.axes[1][start_idx:end_idx]

df_parties_columns = df[parties_columns]
df_parties_columns.head(5)
```

```python
# get the column name of max values in every row
max_party_per_row = df_parties_columns.idxmax(axis=1)
df['winning_party'] = max_party_per_row

df.head(10)
```

```python
df = pd.read_excel('19kneset.xlsx')
df.head(5)
```

```python
df.axes
```

```python
df.rename(columns={'סמל ועדה': 'committee symbol', 'שם ישוב': 'settlement name', 'סמל ישוב': 'settlement symbol', 'בזב': 'suffrages', 'מצביעים':'voters', 'פסולים':'disqualified', 'כשרים':'qualified'}, inplace=True)
print(df.axes)
df.head(5)
```

```python
sums = df.sum(axis=0) # sum all columns, return series
total_suffrages = sums['suffrages']
total_qualified = sums['qualified']
total_voting_percentage = total_qualified / total_suffrages * 100

print("Suffrages: " + str(total_suffrages))
print("Qualified Voters: " + str(total_qualified))
print("voting percentage: " + str(total_voting_percentage))
```

```python
data = sums[5::]
df_parties = pd.DataFrame({'party':data.index, 'voters':data.values})
df_parties['percentage'] = df_parties['voters'] / total_qualified * 100
df_parties = df_parties.sort_values('percentage', ascending=False)
df_parties = df_parties.reset_index(drop=True)
df_parties.head(15)
```

```python
hold_percent = 3.25
hold_number = total_qualified * 3.25 / 100
print("hold percent: " + str(hold_percent))
print("hold number: " + str(hold_number))
```

```python
df_parties = df_parties.loc[df_parties['percentage'] >= 3.25]
print(len(df_parties.index))
df_parties
```

```python
qualified_for_count = df_parties.sum(axis=0)['voters']
mandate_votes = qualified_for_count / 120

df_parties['mandates_float'] = df_parties['voters'] / mandate_votes
df_parties['mandates_int'] = (df_parties['mandates_float']).astype(int)
mandates = df_parties.sum(axis=0)['mandates_int']
missing_mandates = 120 - mandates

#add missing mandates according to the elections committee
missing_mandates = {}

df_parties['mandates_final'] = df_parties['mandates_int']
for k,v in missing_mandates.items():
    df_parties.loc[df_parties['party']==k, ['mandates_final']] += v

df_parties
```
